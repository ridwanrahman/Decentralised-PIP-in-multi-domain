```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
domain1 = pd.read_csv('ABS/time.csv')
generated_keys = pd.read_csv('/home/ridwan/Documents/ABACProjectDomain2/PEP/time_with_generated_keys.csv')
stored_keys = pd.read_csv('/home/ridwan/Documents/ABACProjectDomain2/PEP/time_with_stored_keys.csv')
```

```python
domain1 = domain1[:-1]
domain1 = domain1[:-1]
```

```python
domain1
```

```python
dataFrame = pd.DataFrame(data=domain1);

fig = dataFrame.plot(kind='bar',x="predicate_length", y="time to verify signature", fontsize=20, figsize=(12,8), rot=0);
fig.set_xlabel("Number of Attributes", fontsize=20)
fig.set_ylabel("Time taken to verify signature in seconds", fontsize=20)
fig.set_title("Time to verify signature according to Number of Attributes", fontsize=20)
plt.savefig('time-verify-signature.pdf')
plt.show()
```
